💡`Χρησιμοποιήστε το Column Transformer για τη δημιουργία και την εφαρμογή χωριστών μετασχηματιστών για αριθμητικά και κατηγορικά δεδομένα.`




3.α.  Θα χρησιμοποιήσετε όλα τα χαρακτηριστικά του συνόλου δεδομένων για την εκπαίδευση των ταξινομητών ή θα επιλέξετε κάποια από αυτά;  Θα κάνετε κάποια συνένωση κάποιων χαρακτηριστικών για δημιουργία νέων χαρακτηριστικών για το μοντέλο σας;


We will begin by using almost all features. The feauture we will disregard are ID, as well as the day and year (we will only keep month) !!!!!!!!!!! Ofcourse, RainTomorrow will also be disregarded and the ID column will be stored in a variable in order to "concut" it to the appropriate csv, for the kaggle competition.  
Initially, we will not be adding extra elements to our datasets, exactly. We will, though, get rid of Day and Year in "Date" and only keep "month", which will be converted on a scale of 1-12. 
Also, we will later fill in the missing values...as well as handle the "type-object" values with one-hot encoding

```python
#3a

# Το Column Transformer είναι μια κλάση scikit-learn που χρησιμοποιείται για τη δημιουργία και την εφαρμογή χωριστών μετασχηματιστών για αριθμητικά και κατηγορικά δεδομένα. Για να δημιουργήσουμε μετασχηματιστές πρέπει να καθορίσουμε το αντικείμενο του μετασχηματιστή και να περάσουμε τη λίστα μετασχηματισμών μέσα σε μια πλειάδα μαζί με τη στήλη στην οποία θέλουμε να εφαρμόσουμε τον μετασχηματισμό

# Ο σκοπός του pipeline είναι να "συναρμολογήσει" πολλά βήματα που μπορούν να επικυρωθούν μεταξύ τους, ρυθμίζοντας παράλληλα διαφορετικές παραμέτρους.

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression



df_3 = pd.read_csv('train-val.csv')
df_date_id = df_3.drop(columns=['ID', 'RainTomorrow']) #drop the ID, RainTomorrow columns
df_date_id['Date'] = pd.to_datetime(df_date_id['Date'], errors='coerce').dt.month # keeps only month data and converts it to a 1-12 scale 

print(df_date_id) #this is the df we will be using for preprocessing





# Define the feature lists while excluding 'ID'
num_features = df_date_id.select_dtypes(include=['float64', 'int64']).columns 
cat_features = df_date_id.select_dtypes(include=['object']).columns

# this process doesnt affect df...



# Preprocessing pipelines
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

# The Pipeline object allows chaining together a sequence of transformations. It ensures that all preprocessing steps are applied consistently and in the correct order.
# standard scaler ~ N(0,1)
#  #steps==1 here


categorical_transformer = Pipeline(steps=[
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

#  Converts categorical features into a numerical format suitable for machine learning algorithms by creating binary columns (0s and 1s) for each unique sample of a characteristic in the dataset.
# For example, if a column contains ['red', 'blue', 'green'], it would be transformed into three columns:
# red  blue  green
#  1    0     0
#  0    1     0
#  0    0     1



# Column transformer to apply different transformers to each type of feature
# we won't use this, as we redefine it below
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, num_features),
        ('cat', categorical_transformer, cat_features)
    ]
)

# 'name': A label for the transformer (e.g., 'num' for numerical preprocessing, 'cat' for categorical preprocessing).






# ignore this, down here



# Full pipeline with preprocessing and classifier
# model_pipeline = Pipeline(steps=[
#     ('preprocessor', preprocessor),
#     ('classifier', LogisticRegression())
# ])



# The model_pipeline can now be used for training and prediction
```


3.β. Υπάρχουν απουσιάζουσες τιμές; Γράψτε κατάλληλο κώδικα ώστε να χειριστείτε αυτές τις τιμές.

Yes, the csv file contains nulls. Η αφαίρεση ή αντικατάσταση απουσιάζουσων τιμών από το dataset ειναι η μόνη μορφή προεπεξεργασίας που καλύτερο είναι να γίνει πριν διαχωρίσουμε σε train και test set. Πρακτικά συνήθως χρησιμοποιούμε το μετασχηματιστή “Imputer” του scikit learn που αντικαθιστά κάθε απουσιάζουσα τιμή χαρακτηριστικού με τη μέση τιμή (συνεχείς μεταβλητές) ή την πιο συχνή τιμή (κατηγορικές μεταβλητές) του χαρακτηριστικού στο σύνολο δεδομένων. Η κλάση SimpleImputer λειτουργεί με τη μέθοδο fit_transform που εκτελεί και τις δύο μεθόδους fit() και transform() σε μία μόνο γραμμή.

```python
#3b AND 3c AND 3d AND 3e


from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest,chi2


# Check for missing values
print("Missing values per column:")
print(df_date_id.isnull().sum())

# redefines which columns are numerical vs. object-type
num_features = df_date_id.select_dtypes(include=['float64', 'int64']).columns
cat_features = df_date_id.select_dtypes(include=['object']).columns

# redefine transformers for missing values
# (a, b) represents a=b;

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),  # Replace nulls with mean for numerical columns
    ('scaler', StandardScaler())                  # Scale numerical data
])
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),  # Replace nulls with most frequent value for object-type columns
    ('encoder', OneHotEncoder(handle_unknown='ignore'))    # One-hot encode object-type data
])

# EG, suppose a column contains ['red', 'blue', 'green'], the encoder will create three binary columns (one for each category).

# If the dataset contains a category not seen during training (e.g., 'yellow'), the encoder must decide how to handle this unknown category. now, it ignores it


# handle_unknown='ignore': This parameter instructs the encoder to ignore any unknown categories that might appear in new data when the model is being used. If a category appears in new data that wasn’t in the training set, this setting prevents the code from throwing an error; instead, it handles it gracefully by setting those categories to all-zeros.

# Combine transformations with ColumnTransformer
preprocessor = ColumnTransformer( #ton ksanaorizw
    transformers=[
        ('num', numeric_transformer, num_features),
        ('cat', categorical_transformer, cat_features)
    ]
)



```

 4.α. Εκπαιδεύστε (fit)  και τους  7 ταξινομητές που προαναφέρθηκαν (εμφανίστε τα δείγματα από το train set μαζί με την ετικέτα τους στην είσοδο του κάθε ταξινομητή), χρησιμοποιώντας τις default τιμές για όλες τις παραμέτρους τους.

```python
#Οι μετασχηματιστές αντιπροσωπεύονται από κλάσεις με μέθοδο προσαρμογής (fit), η οποία μαθαίνει παραμέτρους του μοντέλου (π.χ. μέση τιμή και τυπική απόκλιση για κανονικοποίηση) από ένα σύνολο εκπαίδευσης

classifiers = {
    'Naive Bayes': GaussianNB(),
    'KNeighborsClassifier': KNeighborsClassifier(),
    'LogisticRegression': LogisticRegression(),
    'MLP': MLPClassifier(),
    'SVC': SVC(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier()
}

print("X_train dataset:",X_train)
print("Y_train dataset:",y_train)


# The output of the preprocessor is a sparse matrix because of the one-hot encoding step

# Most of the values in the one-hot encoded data are 0s, and only a small number are 1s (for each category in each row). This creates a sparse matrix where most of the entries are zeros.


# Convert sparse matrices to dense format - to ekana epd mou xtypouse kapoio error - to output tou preprocessor einai sparse array, opote prepei na ton kanoume dense

X_train_dense = X_train.toarray() if hasattr(X_train, 'toarray') else X_train
X_val_dense = X_val.toarray() if hasattr(X_val, 'toarray') else X_val

# MLP doesnt converge because of low iterations

# Train each classifier
for name, clf in classifiers.items():
    print(f"\nTraining {name}:")
    clf.fit(X_train_dense, y_train)  # Train the classifier on dense data
    print(f"{name} has been trained.")
```

4.β. Εφαρμόστε στα εκπαιδευμένα μοντέλα από το validation set μόνο τα δείγματα χωρίς την ετικέτα τους (predict).


```python
# Ensure feature scaling if applicable
# If you scaled the training data, the validation data must also be scaled
# Example: X_val_scaled = scaler.transform(X_val)

# store predictions
predictions = {}

# Use each classifier to predict the target labels for the validation set
for name, clf in classifiers.items():
    #print(f"Predicting with {name}:")
    y_pred = clf.predict(X_val_dense)  # Replace X_val with X_val_scaled if scaling was applied
    predictions[name] = y_pred   # Store predictions
    
    #print(f"Predictions for {name} (first 10 samples): {y_pred[:10]}")

```


4.γ. Συγκρίνετε την έξοδο του κάθε μοντέλου σε σχέση με τις αντίστοιχες ετικέτες του validation set και αξιολογήστε την επιδοσή τους χρησιμοποιώντας το F1 score.


```python
from sklearn.metrics import f1_score

# The F1 score ombines two other important metrics — Precision and Recall — into a single score, making it useful when you need a balance between the two.

# Recall: The proportion of true positive predictions out of all actual positive instances in the data.

# Precision: The proportion of true positive predictions out of all positive predictions made by the model.


# Initialize a dictionary to store F1 scores
f1_scores = {}

# Loop through each classifier and evaluate performance
for name, clf in classifiers.items():
    print(f"Evaluating {name}:")
    
    # Predict on validation set
    y_pred = clf.predict(X_val_dense)  # Use X_val_scaled if scaling was applied
    
    # Compute F1 score
    score = f1_score(y_val, y_pred, average='binary')  # we want binary classification
    f1_scores[name] = score
    
    print(f"F1 Score for {name}: {score:.4f}")

# Display all F1 scores
print("\nSummary of F1 Scores:")
for name, score in f1_scores.items():
    print(f"{name}: {score:.4f}")


```
